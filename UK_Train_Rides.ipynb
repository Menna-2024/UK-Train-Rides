```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('railway.csv')
df.head()
```

```python
df.sample(10) #check the data
```

```python
print(df.info()) # data types
```

```python
print(df.describe()) # statistics for numerical column
```

```python
print(df.isnull().sum()) # missing values
```

```python
# Claening the data
# Convert date columns to datetime format
df["Date of Purchase"] = pd.to_datetime(df["Date of Purchase"])
df["Date of Journey"] = pd.to_datetime(df["Date of Journey"])
```

```python
# Convert time columns to time format
time_columns = ["Time of Purchase", "Departure Time", "Arrival Time", "Actual Arrival Time"]
for col in time_columns:
    df[col] = pd.to_datetime(df[col], errors="coerce").dt.time
```

```python
# Fill missing values in 'Actual Arrival Time' with 'Arrival Time' (assuming no delay if missing)
df["Actual Arrival Time"].fillna(df["Arrival Time"], inplace=True)
```

```python
# Fill missing values in 'Reason for Delay' with 'No Delay' for non-delayed journeys
df["Reason for Delay"].fillna("No Delay", inplace=True)
```

```python
most_frequent_railcard = df["Railcard"].mode()[0] # instead of the null mode is used
df["Railcard"].fillna(most_frequent_railcard, inplace=True)
```

```python
# Standardize categorical columns (strip spaces and convert to title case where applicable)
categorical_columns = ["Purchase Type", "Payment Method", "Railcard", "Ticket Class", "Ticket Type",
                        "Departure Station", "Arrival Destination", "Journey Status", "Refund Request"]
df[categorical_columns] = df[categorical_columns].apply(lambda x: x.str.strip().str.title())
```

```python
df.sample(10) # check data after cleaning
```

```python
# Save cleaned dataset
df.to_csv("cleaned_railway_data.csv", index=False)
```
